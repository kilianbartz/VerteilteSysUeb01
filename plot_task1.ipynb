```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./ntp_accuracy_data.csv")
# convert "Local Time" to datetime
df["Local Time"] = pd.to_datetime(df["Local Time"], unit="s")
df
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
xformatter = mdates.DateFormatter('%H:%M')
# plot offset
plt.figure()
plt.plot(df["Local Time"], df["Offset"], label="Offset")
plt.xlabel("UTC Time")
plt.ylabel("Offset (s)")
plt.gcf().axes[0].xaxis.set_major_formatter(xformatter)
plt.legend()
plt.title("NTP Offset")
plt.savefig("ntp_offset_pc.png", dpi=100)
```
